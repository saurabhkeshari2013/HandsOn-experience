```python
# need to handle the outliers, we reduce the impact of outliers.
```

```python
df.columns
```

```python
for i in df.columns:
    if df[i].dtype!='object':
        p1=df[i].quantile(0.01)
        p2=df[i].quantile(0.99)
        df[i][df[i]<p1]=p1
        df[i][df[i]>p2]=p2
```

```python
# standardization
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df_scaled=sc.fit_transform(df_drop)
```

```python
df_scaled
```

```python
df_scaled=pd.DataFrame(df_scaled, columns=df_drop.columns)
```

```python
df_scaled
```

```python
# KMeans model
```

```python
from sklearn.cluster import KMeans
```

```python
# elbow method-- to choose optimal number of cluster
```

```python
k_values=list(range(2,8))
```

```python
ssd=[]
for k in k_values:
    kmodel=KMeans(n_clusters=k, max_iter=150, random_state=23)
    kmodel.fit(df_scaled)
    ssd.append(kmodel.inertia_)
```

```python
ssd
```

```python
k_values
```

```python
# to get elbowplot we plot k_value against ssd
```

```python
plt.plot(k_values, ssd)
plt.show()
```

```python
# here, we get optimal no. of cluster--k=4
```

```python
kmodel=KMeans(n_clusters=4, max_iter=150, random_state=23)
kmodel.fit(df_scaled)
kmodel.labels_
```

```python
#adding new column i.e. "ClusterID" in dataframe
```

```python
df["ClusterID"]=kmodel.labels_
```
